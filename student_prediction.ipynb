```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Load dataset
df = pd.read_csv("Student_score.csv")

# Step 2: Select input (X) and output (y)
X = df[["Hours"]]   # Independent variable (input)
y = df["Score"]     # Dependent variable (output)

# Step 3: Create model
model = LinearRegression()

# Step 4: Train model
model.fit(X, y)

# Step 5: Predict values
predicted = model.predict(X)

# Step 6: Plot graph
plt.scatter(X, y)              # Actual data
plt.plot(X, predicted)         # Predicted line
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Study Hours vs Score Prediction")
plt.show()

# Step 7: Predict new value
print("Score for 5.5 hours:", model.predict([[5.5]])[0])
```
